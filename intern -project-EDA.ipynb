```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("student_info.csv")
print("Dataset Shape:",
df.shape)
print("\nColumn Names: ")
print(df.columns)
print("\nFirst 5 Rows: ")
print(df.head())
print("\nDataset Info:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())
print("\nStatistical Summary:")
print(df.describe())
print("\nAverage Scores: ")
print("Math Score:", df['math_score'].mean())
print("Reading Score:", df['reading_score'].mean())
print("Writing Score:", df['writing_score'].mean())
print("\nFinal Result Count:")
print(df['final_result'].value_counts())
gender_scores = df.groupby('gender') [
['math_score', 'reading_score', 'writing_score']
].mean()
print("\nGender-wise Average Scores:")
print(gender_scores)
attendance_result = df.groupby('final_result')['attendance_rate'].mean()
print("\nAverage Attendance by Final Result:")
print(attendance_result)
top_math = df[['student_id', 'name', 'math_score']].sort_values(
by='math_score', ascending=False
).head(10)
print("\nTop 10 Students in Math:")
print(top_math)
df['final_result'].value_counts().plot(
kind='bar',
title='Pass vs Fail'
)
plt.xlabel('Result')
plt.ylabel('Count')
plt.show()
avg_scores = [
    df['math_score'].mean(),
    df['reading_score'].mean(),
    df['writing_score'].mean()
]
plt.bar(
    ['Math', 'Reading', 'Writing'],
    avg_scores
)
plt.title('Average Scores')
plt.ylabel('Score')
plt.show()
plt.hist(df['attendance_rate'], bins=10)
plt.title('Attendance Rate Distribution')
plt.xlabel('Attendance Rate')
plt.ylabel('Frequency')
plt.show()
```
